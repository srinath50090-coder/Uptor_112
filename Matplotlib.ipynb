# Data Visulaization using Matplotlib

#TATA Motors Dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tatamotors = pd.read_csv("TATAMOTORS.csv",index_col='Date')
```

```python
tatamotors
```

```python
tatamotors.index.dtype
```

```python
tatamotors.index = pd.to_datetime(tatamotors.index,format='%Y-%m-%d')
tatamotors.index.dtype
```

# Line Plot

# Opening Price

```python
plt.figure(figsize=(10,4))
plt.plot(tatamotors.index,tatamotors['Open'],label='TATA Motors Stock Price')
plt.title('Opening Price at Everyday')
plt.xlabel('Date')
plt.ylabel('Open')
plt.legend()
plt.show()
```

# Highest and Lowest Price in a day

```python
fig,ax = plt.subplots(2,1,figsize=(12,10))
ax[0].plot(tatamotors.index,tatamotors['High'],label='High',color='green',linestyle='-',linewidth=2)
ax[0].set_title('Highest Price in a day',)
ax[0].set_xlabel('Date')
ax[0].set_ylabel('High')
ax[0].legend()

ax[1].plot(tatamotors.index,tatamotors['Low'],label='Low',color='red',linestyle='-',linewidth=2)
ax[1].set_title('Lowest Price in a day')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Low')
ax[1].legend()

plt.show()
```

# Histogram

```python
plt.figure(figsize=(10,4))
plt.hist(tatamotors['High'],bins=25,edgecolor='black',color='orange')
plt.title('Highest Price in a day')
plt.xlabel('High')
plt.ylabel('Frequency')
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,8))
ax[0].hist(tatamotors['High'],bins=15,edgecolor='black',color='blue',label='High',orientation='horizontal')
ax[0].set_title('Highest Price in a day')
ax[0].set_ylabel('Highest Price')
ax[0].legend()

ax[1].hist(tatamotors['Low'],bins=15,edgecolor='black',color='red',label='Low',orientation='horizontal')
ax[1].set_title('Lowest Price in a day')
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('Lowest Price')
ax[1].legend()
plt.show()
```

```python
plt.hist(tatamotors['VWAP'],bins=10,edgecolor='black',label='VWAP',histtype='stepfilled',density=True)
plt.title('Volume Weighted Average Price')
plt.xlabel('VWAP')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Bar Chart

# Financial Dataset

```python
finance = pd.read_csv('Financials.csv')
```

```python
finance
```

```python
finance.columns = finance.columns.str.strip()
finance.columns
```

```python
country_profit = finance.groupby('Country')['Profit'].mean()
```

```python
plt.barh(country_profit.index,country_profit,edgecolor='black',color='orange')
plt.title('Total Profit by Country')
plt.xlabel('Profit')
plt.ylabel('Country')
plt.show()
```

```python
canada = finance[finance['Country']=='Canada']
germany = finance[finance['Country']=='Germany']
france = finance[finance['Country']=='France']
mexico = finance[finance['Country']=='Mexico']
us = finance[finance['Country']=='United States of America']
```

```python
canada_product_profit = canada.groupby('Product')['Profit'].mean()
germany_product_profit = germany.groupby('Product')['Profit'].mean()
france_product_profit = france.groupby('Product')['Profit'].mean()
mexico_product_profit = mexico.groupby('Product')['Profit'].mean()
us_product_profit = us.groupby('Product')['Profit'].mean()
```

```python
fig,ax = plt.subplots(3,2,figsize=(14,12))

ax[0][0].bar(canada_product_profit.index,canada_product_profit,edgecolor='black',color='orange')
ax[0][0].set_title('Profit in Canada')
ax[0][0].set_xlabel('Product')
ax[0][0].set_ylabel('Profit')

ax[0][1].bar(germany_product_profit.index,germany_product_profit,edgecolor='black',color='blue')
ax[0][1].set_title('Profit in Germany')
ax[0][1].set_xlabel('Product')
ax[0][1].set_ylabel('Profit')

ax[1][0].bar(france_product_profit.index,france_product_profit,edgecolor='black',color='red')
ax[1][0].set_title('Profit in France')
ax[1][0].set_xlabel('Product')
ax[1][0].set_ylabel('Profit')

ax[1][1].bar(mexico_product_profit.index,mexico_product_profit,edgecolor='black',color='green')
ax[1][1].set_title('Profit in Mexico')
ax[1][1].set_xlabel('Product')
ax[1][1].set_ylabel('Profit')

ax[2][0].bar(us_product_profit.index,us_product_profit,edgecolor='black',color='yellow')
ax[2][0].set_title('Profit in United States')
ax[2][0].set_xlabel('Product')
ax[2][0].set_ylabel('Profit')

plt.tight_layout()
plt.show()
```

# Heat map

```python
medical_student = pd.read_csv('Medical_Students_dataset.csv')
medical_student
```

```python
medical_student_numeric = medical_student[['Height','Weight','BMI','Temperature','Heart Rate','Blood Pressure','Cholesterol']]
```

```python
medical_student_correlation = medical_student_numeric.corr()
medical_student_correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(medical_student_correlation, annot=True, cmap='Reds', fmt=".2f", linewidths=.005)
plt.title('Correlation Heatmap for Medical Student Data')
plt.show()
```

# Box Plot

# Medical Student dataset

```python
plt.figure(figsize=(8,5))
plt.boxplot([medical_student['Heart Rate'],medical_student['BMI'],medical_student['Weight']])
plt.xticks([1,2,3],['Heart Rate','BMI','Weight'])
plt.show()
```

# Pie Chart

```python
carretera = finance[finance['Product']==' Carretera ']
montana = finance[finance['Product']==' Montana ']
paseo = finance[finance['Product']==' Paseo ']
velo = finance[finance['Product']==' Velo ']
vtt = finance[finance['Product']==' VTT ']
amarilla = finance[finance['Product']==' Amarilla ']

carretera_profit = carretera.groupby('Country')['Profit'].mean()
montana_profit = montana.groupby('Country')['Profit'].mean()
paseo_profit = paseo.groupby('Country')['Profit'].mean()
velo_profit = velo.groupby('Country')['Profit'].mean()
vtt_profit = vtt.groupby('Country')['Profit'].mean()
amarilla_profit = amarilla.groupby('Country')['Profit'].mean()
```

```python
fig, ax = plt.subplots(3,2,figsize=(10,10))

ax[0][0].pie(carretera_profit,labels=carretera_profit.index,autopct='%1.2f%%')
ax[0][0].set_title('Carretera Profit')

ax[0][1].pie(montana_profit,labels=montana_profit.index,autopct='%1.2f%%',explode=[0,0,0.1,0,0])
ax[0][1].set_title('Montana Profit')

ax[1][0].pie(paseo_profit,labels=paseo_profit.index,autopct='%1.2f%%',explode=[0.1,0,0,0,0])
ax[1][0].set_title('Paseo Profit')

ax[1][1].pie(velo_profit,labels=velo_profit.index,autopct='%1.2f%%',explode=[0,0,0.05,0,0])
ax[1][1].set_title('Velo Profit')

ax[2][0].pie(vtt_profit,labels=vtt_profit.index,autopct='%1.2f%%',explode=[0.05,0.05,0.05,0.05,0.05])
ax[2][0].set_title('VTT Profit')

ax[2][1].pie(amarilla_profit,labels=amarilla_profit.index,autopct='%1.2f%%',explode=[0.05,0.05,0.05,0.05,0.05])
ax[2][1].set_title('Amarilla Profit')

plt.tight_layout()
plt.show()
```

# Stack Plot

```python
us_trade = pd.read_csv('US_trade.csv')
```

```python
us_trade
```

```python
plt.figure(figsize=(14, 7))

# Prepare data for the stack plot
countries = ['China', 'Japan', 'Mexico', 'Canada', 'Germany', 'France', 'India', 'UK', 'Taiwan']

# Plotting the stack plot
plt.stackplot(us_trade['Year'], us_trade[countries].T, labels=countries, alpha=0.8)

plt.title('US Trade Proportions by Country Over Years', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Trade Proportion (%)')
plt.xticks(us_trade['Year'][::1],rotation=45)
plt.legend(loc='upper right')
plt.show()
```
